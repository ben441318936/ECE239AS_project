```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
deep_eval_data = np.load("./TD3_kuka_random_ball_random_target_deep/eval_logs/evaluations.npz")
shallow_eval_data = np.load("./TD3_kuka_random_ball_random_target_shallow/eval_logs/evaluations.npz")
```

```python
shallow_results = shallow_eval_data["results"]
deep_results = deep_eval_data["results"]
```

```python
shallow_ep_rewards = shallow_results[:,:,0]
shallow_ep_rewards = np.mean(shallow_ep_rewards,axis=1)

deep_ep_rewards = deep_results[:,:,0]
deep_ep_rewards = np.mean(deep_ep_rewards,axis=1)
```

```python
eps = np.arange(0,1000)*1000
```

```python
plt.figure(figsize=(5,5))
plt.semilogy(eps,shallow_ep_rewards)
plt.semilogy(eps,deep_ep_rewards)
plt.plot()
plt.xlabel("Training timesteps",fontsize=14)
plt.xticks(fontsize=14)
plt.ylabel("Mean evaluation episodic reward",fontsize=14)
plt.yticks(fontsize=14)
plt.legend(["TD3 shallow","TD3 deep"],fontsize=12, loc="lower right")
plt.tight_layout()
plt.savefig("./TD3_shallow_vs_deep_random_ball_random_target_curve.png",dpi=800)
```
